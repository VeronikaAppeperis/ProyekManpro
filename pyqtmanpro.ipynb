```python
import sys
from PyQt5.QtWidgets import QApplication, QMainWindow, QWidget, QPushButton, QTableWidget, QTableWidgetItem, QHeaderView,  QLabel, QLineEdit, QVBoxLayout
from PyQt5.uic import loadUi
from PyQt5.QtCharts import QChart, QChartView, QAreaSeries, QPieSeries, QLineSeries, QBarSeries, QBarSet, QBarCategoryAxis
```

```python
app = QApplication(sys.argv)
```

```python
# import aplikasi pyqt

window = QWidget()
window.setWindowTitle('PyQt Application')
window.show()

sys.exit(app.exec_())
```

```python

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        loadUi('mainwindow.ui', self)
        self.show()

app = QApplication(sys.argv)
window = MainWindow()
sys.exit(app.exec_())
```

```python
# membuat header

app = QApplication([])
window = QMainWindow()


table = QTableWidget(window)
table.setColumnCount(3)
table.setRowCount(2)
table.setItem(0, 0, QTableWidgetItem('Main'))
table.setItem(0, 1, QTableWidgetItem('Sort'))
table.setItem(0, 2, QTableWidgetItem('Item 3'))
table.setItem(1, 0, QTableWidgetItem('Item 4'))
table.setItem(1, 1, QTableWidgetItem('Item 5'))
table.setItem(1, 2, QTableWidgetItem('Item 6'))

header = table.horizontalHeader()
header.setSectionResizeMode(QHeaderView.Stretch)

window.setCentralWidget(table)
window.show()
app.exec_()
```

```python
#untuk membuayt inputan

layout = QVBoxLayout()

label1 = QLabel('Name:', window)
input1 = QLineEdit(window)
label2 = QLabel('Age:', window)
input2 = QSpinBox(window)

layout.addWidget(label1)
layout.addWidget(input1)
layout.addWidget(label2)
layout.addWidget(input2)

window.setLayout(layout)
window.show()
app.exec_()
```
